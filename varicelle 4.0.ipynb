```python
def pos():
    position=[]#liste de positions
    
    for i in range(7):
        for j in range(7):
            position.append([i,j])
            np.random.shuffle(position)
    return (position)
```

```python
def cases_vides():
    vide=[]
    position=pos()
    i=0
    while (i<24):
        vide.append(position[i])
        i=i+1
        
    return(vide)
    
```

```python
def voisin():
    elements=[]
    voisins={}
    for i in range(7):
        for j in range(7):
            elements.append((i,j))
            if (i,j)==(0,0):
                voisins[(i,j)]=[(i+1,j),(i,j+1)]
            elif (i,j)==(0,6):
                voisins[(i,j)]=[(i,j-1),(i+1,j)]
            elif (i,j)==(6,0):
                voisins[(i,j)]=[(i-1,j),(i,j+1)]
            elif (i,j)==(6,6):
                voisins[(i,j)]=[(i,j-1),(i-1,j)]
                
            elif (i==0 and j>=1 and j<=5):
                voisins[(i,j)]=[(i+1,j),(i,j+1),(i,j-1)]
                
                
            elif (i==6 and j>=1 and j<=5):
                voisins[(i,j)]=[(i-1,j),(i,j+1),(i,j-1)]
                
                
            elif (j==0 and i>=1 and i<=5):
                voisins[(i,j)]=[(i+1,j),(i,j+1),(i-1,j)]
            
            elif (j==6 and i>=1 and i<=5):
                voisins[(i,j)]=[(i+1,j),(i-1,j),(i,j-1)]
            else:
                voisins[(i,j)]=[(i+1,j),(i-1,j),(i,j-1),(i,j+1)]
    return(voisins)
```

```python
def _cases_vides(mat):
    cases=cases_vides()
    for k in cases:
        [a,b]=k
        mat[a][b]=0
    return mat
  
def initialisation_matrice():
    matrice = np.ones([7,7])
    matrice = _cases_vides(matrice)
    cases_un = [list(x) for x in np.argwhere(matrice == 1)]
    deux_r=random.choice(cases_un)
    i=deux_r[0]
    j=deux_r[1]

    matrice[i][j]=2

    return matrice
```

```python
matrice2 = initialisation_matrice()
vraideplacement(matrice2)
```
